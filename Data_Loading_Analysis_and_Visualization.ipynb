import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Display  plots in the notebook and seaborn style

```python
%matplotlib inline
sns.set_style("whitegrid")
```

```python
#load the dataset
try:
    # Example with seaborn's iris dataset (replace with pd.read_csv('your_file.csv') for your own data)
    df = sns.load_dataset('iris')
    print("✅ Dataset loaded successfully!\n")
except FileNotFoundError:
    print("❌ File not found. Please check the file path.")
except Exception as e:
    print(f"❌ An error occurred while loading the dataset: {e}")
```

```python
#Display first few rows
print("First 5 rows of the dataset:")
display(df.head())

#Check structure, datatypes, and missing values
print("\nDataset Info:")
df.info()


print("\nMissing Values:")
print(df.isnull().sum())

# Clean missing values (if any)
if df.isnull().sum().any():
    df = df.dropna()  # Or you can fillna() depending on context
    print("\nMissing values handled.")

# 
# Task 2: Basic Data Analysis
# 

# 1.Basic statistics
print("\nBasic Statistics:")
display(df.describe())

# 2.Group by categorical column (species) and compute mean of numerical columns
print("\nAverage measurements per species:")
grouped = df.groupby('species').mean(numeric_only=True)
display(grouped)

# 3.Observations / Patterns to be found 
# 
# Task 3: Data Visualization
# 

# 1.Line chart (Example: cumulative mean sepal length to simulate trend)
plt.figure(figsize=(7,4))
df['sepal_length'].expanding().mean().plot(color='blue', linewidth=2)
plt.title("Trend of Cumulative Mean Sepal Length")
plt.xlabel("Sample Index")
plt.ylabel("Cumulative Mean Sepal Length (cm)")
plt.show()

# 2. Bar chart (Average petal length per species)
plt.figure(figsize=(7,4))
grouped['petal_length'].plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()

# 3Histogram (Distribution of sepal width)
plt.figure(figsize=(7,4))
plt.hist(df['sepal_width'], bins=15, color='purple', edgecolor='black')
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()

# 4.Scatter plot (Sepal length vs Petal length)
plt.figure(figsize=(7,4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df, s=80)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title='Species')
plt.show()

# 
# Findings & Conclusion here
```
